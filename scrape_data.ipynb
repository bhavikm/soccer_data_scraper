```python
soup = BeautifulSoup(tree.content, 'html.parser')
```

```python
script = str(soup.findAll('script')[-1])

pattern = "'series':\[(.*)\]"
extract = re.search(pattern, script)

if extract:
 found = extract.group(1)

print(found)
```
